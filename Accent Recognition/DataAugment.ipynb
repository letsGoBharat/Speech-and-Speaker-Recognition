```
def modif_speed(filename,integer,percentage,son):
    
    filename = filename +'_'+str(integer) +'.wav'
    debut = time ()
    
    freq = son.getframerate()
    frames = son.getnframes()
    print(freq)
    son_final = wave.open(filename,'wb')
    
    freq_final = int(percentage*freq/100)
    print(freq_final)
    
    son_final.setparams((son.getnchannels(),son.getsampwidth(),freq_final,frames,'NONE','not compressed'))
    son_final.writeframes(son.readframes(frames))
    
    son_final.close()
    
    fin = time()
    print(fin-debut)
```
